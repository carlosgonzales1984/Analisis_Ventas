```python
# 1. Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Cargar datos
df = pd.read_csv('/kaggle/input/supermarket-sales/supermarket_sales_150 - Sheet1.csv') 
```

```python
# 3. Vista general
df.info()
df.describe()
```

```python
# 4. Limpieza (si aplica)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Ventas por línea de producto
ventas_producto = df.groupby('Product Category')['Total Sales'].sum().sort_values()

plt.figure(figsize=(8,4))
ventas_producto.plot(kind='barh', color='orange')
plt.title('Ventas por Línea de Producto')
plt.xlabel('Total vendido')
plt.tight_layout()
plt.show()
```
